```python
# Import required libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import sklearn
from sklearn.neural_network import MLPClassifier #We will use this to create our model
from sklearn.neural_network import MLPRegressor

# Import necessary modules
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('diabetes.csv') 
print(df.shape)
df.describe().transpose()
```

```python
target_column = ['Outcome']   #creates an object of the target variable called 'target_column'.
predictors = list(set(list(df.columns))-set(target_column)) # Here we created predictors we selected all columns and then subtracted target column ie Outcome
df[predictors] = df[predictors]/df[predictors].max() # This normalizes the predictors.
df.describe().transpose()
```

```python
X = df[predictors].values
y = df[target_column].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=40)
print(X_train.shape); print(X_test.shape)
```

```python
#we will build the neural network model using the scikit-learn library's estimator object, 
#'Multi-Layer Perceptron Classifier'
from sklearn.neural_network import MLPClassifier

# Now we will instantiates the model with the 'hidden_layer_sizes' argument set to three layers, which has the same number
# of neurons as the count of features in the dataset. We will also select 'relu' as the activation function and
# 'adam' as the solver for weight optimization.
mlp = MLPClassifier(hidden_layer_sizes=(8,8,8), activation='relu', solver='adam', max_iter=500)

#Train The model
mlp.fit(X_train,y_train)

#predict the data
predict_train = mlp.predict(X_train)   #This will predict values of y from training data
predict_test = mlp.predict(X_test)    #This will predict values of y from testing data
```

```python
#This is performance of model on training data
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_train,predict_train)) # Testing accuracy of - actual values (y_train) vs values predicted using trainging data (predict_train)
print(classification_report(y_train,predict_train))
```

```python
#This is performance of model on training data
print(confusion_matrix(y_test,predict_test)) # Testing accuracy of - actual values (y_test) vs values predicted using testing data (predict_test)
print(classification_report(y_test,predict_test))
```
